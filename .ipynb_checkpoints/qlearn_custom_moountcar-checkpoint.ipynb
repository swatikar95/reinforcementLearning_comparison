```python
import gym
from gym import spaces
import numpy as np
```

```python
class CustomMountainCarEnv(gym.Env):
    def __init__(self):
        super(CustomMountainCarEnv, self).__init__()

        # Define the observation space
        self.observation_space = spaces.Box(
            low=np.array([-1.2, -0.07]),  # Minimum position and velocity
            high=np.array([0.6, 0.07]),   # Maximum position and velocity
            dtype=np.float32
        )

        # Define the action space
        self.action_space = spaces.Box(
            low=np.array([-1.0]),  # Minimum action value
            high=np.array([1.0]),  # Maximum action value
            dtype=np.float32
        )

        # Define the initial position and velocity
        self.start_position = -0.8
        self.start_velocity = 0.0

        # Define the goal position
        self.goal_position = 0.2

        # Define the maximum and minimum position values
        # self.max_position = 0.6
        # self.min_position = -1.2

        # Initialize the state
        self.state = None
       

        self.max_steps = 200  # Maximum number of steps
        self.current_step = 0  # Current step count
        
        # self.position_history = []  # Track position changes


    def step(self, action):
        # Unpack the action
        force = action[0]

        # Clip the force within the action space limits
        force = np.clip(force, self.action_space.low, self.action_space.high)

        # Update the velocity and position based on the force
        self.state[1] += force 
        self.state[1] = np.clip(self.state[1], self.observation_space.low[1], self.observation_space.high[1])
        self.state[0] += self.state[1]
        self.state[0] = np.clip(self.state[0], self.observation_space.low[0], self.observation_space.high[0])
        
        self.current_step += 1

        # Determine if the car has reached the goal
        if self.current_step >= self.max_steps:
            done = True
            reward = -1.0
        else:
            done = bool(self.state[0] >= self.goal_position)
            if done:
                reward = 100.0
            else:
                reward = -1.0

        # self.position_history.append(self.state[0])
        # Calculate the reward
        # if done:
        #     reward = 100.0
        # else:
        #     reward = -1.0

        return self.state, reward, done, {}

    def reset(self):
        # Initialize the state with random position and zero velocity
        self.state = np.array([self.start_position, self.start_velocity])
        self.current_step = 0  # Reset the step count

        # self.position_history = []  # Reset position history
        
        return self.state
    
    # def print_episode_history(self):
    #     for episode in self.episode_history:
    #         print(episode)

```
